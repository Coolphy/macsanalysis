## CeCoIn5 Data Analysis 

### Basic Information
Experiment MACS, NIST, 1.25-1.31,2018.    
All measurements were performed with magnetic field B = 4T. Measured at 0.1K (base temperature) and 2.5K (high temperature). Be filters are used both before and after the sample.


```python
import sys, os
directory = "E:\Dropbox\Research\experiment data\CeCoIn5_MACS_20180125\macsanalysis"
os.chdir(directory)
sys.path.append(directory)
```

Projection data generated from Dave are imported. ***ltData*** stands for low temperature (T=0.1K) data, and ***htData*** stands for high temperature (T=2.5K) data. ***subData*** stands for bankground subtracted data generated by Dave empty can subtraction method. All analysis below depend on DataClass. 

```python
from DataClass import *
ltFilename = "CeCoIn5_0p1K_4T.txt"
ltData = MACSData(ltFilename) # low temperature data.
htFilename = "CeCoIn5_2p5K_4T.txt"
htData = MACSData(htFilename) # high temperature data.
subFilename = "CeCoIn5_subtraction_4T.txt" 
subData = MACSData(subFilename) # low temperature - high temperature data.
```

### Raw Data visualization, constant E slice.

```python
# -------HK plane, unfolded data--------
E = 0.3
bin_ax1 = [-1, 0.02, 1]
bin_ax2 = [-2, 0.02, 0.5]
bin_ax3 = [E - 0.01, E + 0.01]

ltfigure = ltData.plot(view_ax=12, bin_ax1=bin_ax1, bin_ax2=bin_ax2, bin_ax3=bin_ax3, foldmode=0, 
                       plotflag=True, title='T = 0.1K, E = '+str(E)+'meV', clim=(0,2), xlabel='HH', ylabel='L')
htfigure = htData.plot(view_ax=12, bin_ax1=bin_ax1, bin_ax2=bin_ax2, bin_ax3=bin_ax3, foldmode=0, 
                       plotflag=True, title='T = 2.5K, E = '+str(E)+'meV', clim=(0,2), xlabel='HH', ylabel='L')
subfigure = subtraction(ltfigure, htfigure)
subfigure.plot(title='T = 0.1 - 2.5K, E = '+str(E)+'meV',clim=(-0.5,1), xlabel='HH', ylabel='L')
```

### Low Temperature - High Temperature Data, folded data, constant E slice.

```python
# -------HL plane, folded data--------
view_ax = 12
E = 0.3
bin_ax1 = [-0.2, 0.02, 1]
bin_ax2 = [-0.2, 0.02, 2]
bin_ax3 = [E - 0.01, E + 0.01]

ltfigure = ltData.plot(view_ax=view_ax, bin_ax1=bin_ax1, bin_ax2=bin_ax2, bin_ax3=bin_ax3, foldmode=12, plotflag=False, 
                       title='T = 0.1K, E = '+str(E)+'meV', clim=(0,2), xlabel='HH', ylabel='L')
htfigure = htData.plot(view_ax=view_ax, bin_ax1=bin_ax1, bin_ax2=bin_ax2, bin_ax3=bin_ax3, foldmode=12, plotflag=False, 
                       title='T = 2.5K, E = '+str(E)+'meV', clim=(0,2), xlabel='HH', ylabel='L')
subfigure = subtraction(ltfigure, htfigure)
subfigure.plot(title='T = 0.1 - 2.5K, E = '+str(E)+'meV',clim=(-0.5,1), xlabel='HH', ylabel='L')
```

### Low Temperature - High Temperature Data, folded data, constant L slice.

```python
# -------HE plane, folded data--------
bin_ax1 = [-0.2, 0.02, 1]
bin_ax2 = [0.3, 0.7]
bin_ax3 = [0.2, 0.1, 1]

ltfigure = ltData.plot(view_ax=13, bin_ax1=bin_ax1, bin_ax2=bin_ax2, bin_ax3=bin_ax3, foldmode=12, plotflag=False, 
                       title='T = 0.1K, L = '+str(bin_ax2), clim=(0,2), xlabel='HH', ylabel='E')
htfigure = htData.plot(view_ax=13, bin_ax1=bin_ax1, bin_ax2=bin_ax2, bin_ax3=bin_ax3, foldmode=12, plotflag=False, 
                       title='T = 2.5K, L = '+str(bin_ax2), clim=(0,2), xlabel='HH', ylabel='E')
subfigure = subtraction(ltfigure, htfigure)
subfigure.plot(title='T = 0.1 - 2.5K, L = '+str(bin_ax2),clim=(-0.5,1), xlabel='HH', ylabel='E')
```

### Low Temperature - High Temperature Data, folded data, constant HH slice

```python
# -------LE plane, folded data--------
bin_ax1 = [0.4,0.6]
bin_ax2 = [0, 0.02, 1]
bin_ax3 = [0.2, 0.1, 1]

ltfigure = ltData.plot(view_ax=23, bin_ax1=bin_ax1, bin_ax2=bin_ax2, bin_ax3=bin_ax3, foldmode=12, plotflag=False, 
                       title='T = 0.1K, HH = '+str(bin_ax1), clim=(0,2), xlabel='L', ylabel='E')
htfigure = htData.plot(view_ax=23, bin_ax1=bin_ax1, bin_ax2=bin_ax2, bin_ax3=bin_ax3, foldmode=12, plotflag=False, 
                       title='T = 2.5K, HH = '+str(bin_ax1), clim=(0,2), xlabel='L', ylabel='E')
subfigure = subtraction(ltfigure, htfigure)
subfigure.plot(title='T = 0.1 - 2.5K, HH = '+str(bin_ax1),clim=(-0.5,1), xlabel='L', ylabel='E')
```

### Energy Scan, folded data

```python
bin_ax1 = [0.45, 0.55]
bin_ax2 = [0.45, 0.55]

ltEScan = ltData.plot(view_ax=3,bin_ax1=bin_ax1,bin_ax2=bin_ax2,bin_ax3=[0.2,0.1,1],foldmode=12,plotflag=False)
htEScan = htData.plot(view_ax=3,bin_ax1=bin_ax1,bin_ax2=bin_ax2,bin_ax3=[0.2,0.1,1],foldmode=12,plotflag=False)
subEScan = subtraction(ltEScan, htEScan, plotmode=1, marker='x',linestyle='--',legend=['0.1K','2.5K'],
                       title='Energy Scan B = 4T, HH:'+str(bin_ax1)+", L:"+str(bin_ax2),
                       xlabel='E (meV)',ylabel='Intensity')
subEScan.plot(marker='x',linestyle='--',title='Energy Scan B = 4T, HH:'+str(bin_ax1)+", L:"+str(bin_ax2),
              legend=['0.1 - 2.5K'],
              xlabel='E (meV)',ylabel='Intensity')
```
